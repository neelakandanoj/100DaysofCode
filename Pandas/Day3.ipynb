```python
from bs4 import BeautifulSoup
```

```python
Out=BeautifulSoup(response.text,'html.parser')
```

```python
Out
```

```python
l=Out.find_all('table',{'class':'wikitable'})
```

```python
l[0]
```

```python
import pandas as pd
a=pd.read_html(str(l[0]))
a
```

```python
df=a[0]
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns='Source', inplace=True)
```

```python
df.head()
```

```python
df['Average weight']
```

```python
a='60.7 kg (133.8 lb)'
```

```python
a.split('(')[0]
```

```python
import re
float(re.findall('\d+\.\d+',a.split('(')[0])[0])
```

```python
df['Averafe weight in kg']=df['Average weight'].apply(lambda x: float(re.findall('\d+\.\d+',x.split('(')[0])[0]))
```

```python
df.head()
```

```python
df['Averafe weight in lb']=df['Average weight'].apply(lambda x: float(re.findall('\d+\.\d+',x.split('(')[1])[0]))
```
